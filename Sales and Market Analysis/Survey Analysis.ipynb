## Share weights correction

Now lets make sample sizes closer to the population and calculate weights to multiply

```python
survey_share_2019 = (
    questions_df_2019["country."].reset_index().groupby(["country"]).count()
)

survey_share_2019["survey_share"] = (
    survey_share_2019.iloc[:, 0] / questions_df_2019["country."].shape[0]
)

survey_share_2019 = survey_share_2019.reset_index()

survey_share_2019.columns = ["country", "sample size", "survey share"]

survey_share_2019.head()
```

```python
questions_df_2019["country."] = pd.merge(
    questions_df_2019["country."],
    population[["country", "population", "world share"]],
    on="country",
    how="left",
)
questions_df_2019["country."] = pd.merge(
    questions_df_2019["country."], survey_share_2019, on="country", how="left"
)

questions_df_2019["country."]["weight"] = (
    questions_df_2019["country."]["world share"]
    / questions_df_2019["country."]["survey share"]
)

questions_df_2019["country."].head()
```

```python
survey_share_2020 = (
    questions_df_2020["country."].reset_index().groupby(["country"]).count()
)

survey_share_2020["survey_share"] = (
    survey_share_2020.iloc[:, 0] / questions_df_2020["country."].shape[0]
)

survey_share_2020 = survey_share_2020.reset_index()

survey_share_2020.columns = ["country", "sample size", "survey share"]

survey_share_2020.head()
```

```python
questions_df_2020["country."] = pd.merge(
    questions_df_2020["country."],
    population[["country", "population", "world share"]],
    on="country",
    how="left",
)
questions_df_2020["country."] = pd.merge(
    questions_df_2020["country."], survey_share_2020, on="country", how="left"
)

questions_df_2020["country."]["weight"] = (
    questions_df_2020["country."]["world share"]
    / questions_df_2020["country."]["survey share"]
)

questions_df_2020["country."].head()
```

## Analyse C++ IDE

```python
us_indices_2019 = (
    questions_df_2019["country."]
    .index[questions_df_2019["country."]["country"] == "United States"]
    .tolist()
)

cpp_ide_with_countries_2019 = pd.merge(
    questions_df_2019["cpp_ide."],
    questions_df_2019["country."][["country", "weight"]],
    left_index=True,
    right_index=True,
)

cpp_ide_with_countries_2019.head()
```

```python
questions_df_2019["cpp_ide."].columns.unique()
```

```python
ww_ide_stat_2019 = {}

for ide in questions_df_2019["cpp_ide."].columns:
    ww_ide_stat_2019[ide] = (
        cpp_ide_with_countries_2019.groupby([ide, "country"])
        .agg({"weight": ["sum"]})
        .sum()[0]
    )

ww_cpp_ide_2019 = pd.DataFrame.from_dict(
    ww_ide_stat_2019, orient="index", columns=["value"]
)

ww_cpp_ide_2019.index.names = ["ide"]

ww_cpp_ide_2019.head()
```

```python
questions_df_2019["c_ide."].iloc[:, 0].unique()
```

```python
ww_c_ide_2019 = (
    pd.merge(
        questions_df_2019["c_ide."],
        questions_df_2019["country."][["weight"]],
        left_index=True,
        right_index=True,
    )
    .groupby(["ide"])
    .agg({"weight": [("value", "sum")]})
)

ww_c_ide_2019.columns = ww_c_ide_2019.columns.get_level_values(1)

ww_c_ide_2019 = ww_c_ide_2019.rename(index={"Vi / Vim": "Vi/Vim"})

ww_c_ide_2019.head()
```

```python
ww_c_cpp_ide_2019 = pd.merge(
    ww_cpp_ide_2019, ww_c_ide_2019, left_index=True, right_index=True, how="left"
)

ww_c_cpp_ide_2019["value"] = ww_c_cpp_ide_2019["value_x"] + ww_c_cpp_ide_2019[
    "value_y"
].fillna(0)

ww_c_cpp_ide_2019.columns = [
    "cpp_responses_value",
    "c_responses_value",
    "c_cpp_responses_value",
]

ww_c_cpp_ide_2019[
    [
        "% of all respondents for cpp ide",
        "% of all respondents for c ide",
        "% of all respondents for c/cpp ide",
    ]
] = (
    ww_c_cpp_ide_2019 / ww_c_cpp_ide_2019.sum() * 100
)

ww_c_cpp_ide_2019.head()
```

```python
ww_c_cpp_ide_2019.sort_values("% of all respondents for c/cpp ide")[
    "% of all respondents for c/cpp ide"
].plot.barh(legend=False)

plt.gcf().text(0.8, 0.02, "*who use C/C++")
plt.xlabel("% of all respondents*")
plt.ylabel("")
plt.title("Main C/C++ IDE 2019 Worldwide")

plt.savefig("images/main_c_cpp_ide_2019_ww.png", bbox_inches="tight");
```

# Analyse Usage of Version Control Systems within Visual Studio Code users

```python
vcs_how_with_countries_2019 = pd.merge(
    questions_df_2019["vcs_how."].loc[code_indices_2019],
    questions_df_2019["country."][["country", "weight"]],
    left_index=True,
    right_index=True,
)

vcs_how_with_countries_2019.head()
```

```python
vcs_how_stat_2019 = {}

for vcs_how in questions_df_2019["vcs_how."].columns:
    vcs_how_stat_2019[vcs_how] = (
        vcs_how_with_countries_2019.groupby([vcs_how, "country"])
        .agg({"weight": ["sum"]})
        .sum()[0]
    )

ww_vcs_how_2019 = pd.DataFrame.from_dict(
    vcs_how_stat_2019, orient="index", columns=["value"]
)

ww_vcs_how_2019["% of all respondents*"] = (
    ww_vcs_how_2019["value"]
    / vcs_how_with_countries_2019.dropna(
        axis=0, how="all", subset=questions_df_2019["vcs_how."].columns
    )["weight"].sum()
    * 100
)

ww_vcs_how_2019 = ww_vcs_how_2019.drop(columns=["value"])

ww_vcs_how_2019.index.names = ["vcs_how"]

ww_vcs_how_2019.head(10)
```
